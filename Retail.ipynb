# 3. Data Visualisation

- Explore the distribution of key variables like Sales, Profit, Quantity, and Discount.

```python
df1[['Sales', 'Quantity', 'Discount', 'Profit']].describe()
```

```python
# Drawing Heatmap to identify correlation between numerical columns:
plt.figure(figsize=(8,6))
correlation=df1.corr()
sns.heatmap(correlation,annot=True,cmap="RdYlGn")
plt.show()
```

```python
plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
sns.boxplot(x=df1['Sales'])
plt.title('Boxplot of Sales')

plt.subplot(2, 2, 2)
sns.boxplot(x=df1['Quantity'])
plt.title('Boxplot of Quantity')

plt.subplot(2, 2, 3)
sns.boxplot(x=df1['Discount'])
plt.title('Boxplot of Discount')

plt.subplot(2, 2, 4)
sns.boxplot(x=df1['Profit'])
plt.title('Boxplot of Profit')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
sns.histplot(df1['Sales'], bins=30, kde=True)
plt.title('Distribution of Sales')

plt.subplot(2, 2, 2)
sns.histplot(df1['Quantity'], bins=30, kde=True)
plt.title('Distribution of Quantity')

plt.subplot(2, 2, 3)
sns.histplot(df1['Discount'], bins=30, kde=True)
plt.title('Distribution of Discount')

plt.subplot(2, 2, 4)
sns.histplot(df1['Profit'], bins=30, kde=True)
plt.title('Distribution of Profit')

plt.tight_layout()
plt.show()
```

-  Analyze sales and profit by product categories.

-  Identify categories with high and low performance.

```python
plt.figure(figsize=(8, 4))
sns.barplot(x='Category', y='Sales', data=df1, estimator=sum, ci=None)
plt.title('Total Sales by Category')
plt.show()

plt.figure(figsize=(8, 4))
sns.barplot(x='Category', y='Profit', data=df1, estimator=sum, ci=None)
plt.title('Total Profit by Category')
plt.show()
```

-  Explore sales and profit by customer segments (e.g., by region, by customer segment).

```python
plt.figure(figsize=(8, 6))
sns.barplot(x='Region', y='Sales', data=df1, estimator=sum, ci=None)
plt.title('Total Sales by Region')
plt.show()

plt.figure(figsize=(8, 6))
sns.barplot(x='Region', y='Profit', data=df1, estimator=sum, ci=None)
plt.title('Total Profit by Region')
plt.show()
```

-  Analyze the impact of discounts on sales and profitability.

```python
# Impact of Discount on Profit
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Discount', y='Profit', data=df1)
plt.title('Impact of Discount on Profit')
plt.show()
```
